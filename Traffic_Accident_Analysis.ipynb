```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample traffic accident dataset
data = {
    'Weather': ['Clear', 'Rain', 'Fog', 'Clear', 'Rain', 'Fog', 'Clear', 'Rain'],
    'Road_Condition': ['Dry', 'Wet', 'Wet', 'Dry', 'Wet', 'Wet', 'Dry', 'Wet'],
    'Time_of_Day': ['Morning', 'Night', 'Evening', 'Afternoon', 'Night', 'Morning', 'Evening', 'Night'],
    'Severity': [2, 4, 3, 1, 4, 3, 2, 5]
}

# Create dataframe
df = pd.DataFrame(data)

# Display dataset
print("Traffic Accident Dataset:")
print(df)

# -----------------------------
# Weather Condition Analysis
# -----------------------------
plt.figure(figsize=(6,4))

sns.countplot(x='Weather', data=df)

plt.title("Accidents by Weather Condition")

plt.show()

# -----------------------------
# Road Condition Analysis
# -----------------------------
plt.figure(figsize=(6,4))

sns.countplot(x='Road_Condition', data=df)

plt.title("Accidents by Road Condition")

plt.show()

# -----------------------------
# Time of Day Analysis
# -----------------------------
plt.figure(figsize=(8,4))

sns.countplot(x='Time_of_Day', data=df)

plt.title("Accidents by Time of Day")

plt.show()

# -----------------------------
# Severity Distribution
# -----------------------------
plt.figure(figsize=(6,4))

sns.histplot(df['Severity'], bins=5)

plt.title("Accident Severity Distribution")

plt.xlabel("Severity")

plt.ylabel("Frequency")

plt.show()

# -----------------------------
# Accident Hotspot Visualization
# -----------------------------
plt.figure(figsize=(6,4))

sns.scatterplot(
    x=[1,2,3,4,5,6,7,8],
    y=[2,3,1,5,4,6,5,7],
    hue=df['Severity'],
    size=df['Severity'],
    sizes=(50,300)
)

plt.title("Accident Hotspots Visualization")

plt.show()
```
